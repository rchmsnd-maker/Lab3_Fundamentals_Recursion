```python
# Initial Power: 4 + 11 = 15
def signal_shutdown(power, count=0):
    if power == 0:
        print("Signal Strength: 0 - Shutdown Complete")
        return count
    print(f"Current Signal Strength: {power}")
    return signal_shutdown(power - 1, count + 1)

total_calls = signal_shutdown(15)
```
